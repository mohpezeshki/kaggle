```python
df = pd.read_csv("/kaggle/input/playground-series-s4e6/train.csv" , index_col="id")
X = df.drop(["Target"] , axis=1)
y = df["Target"]
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_transformed_df = le.fit_transform(y)
y_transformed_df = pd.Series(y_transformed_df , name=y.name , index=y.index)
y_transformed_df
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y_transformed_df, test_size = 0.2, random_state = 1)
```

```python
from sklearn.preprocessing import StandardScaler
sta = StandardScaler()
X_train = sta.fit_transform(X_train)
X_test = sta.transform(X_test)
```

```python
from xgboost import XGBClassifier

my_model = XGBClassifier()
# Add silent=True to avoid printing out updates with each cycle
my_model.fit(X_train, y_train, verbose=False)
```

```python
from sklearn.metrics import accuracy_score

y_pred = my_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)
```

```python
df_test = pd.read_csv("/kaggle/input/playground-series-s4e6/test.csv" , index_col="id")
X_test = sta.transform(df_test)
y_pred = my_model.predict(X_test)
y_pred = le.inverse_transform(y_pred)

result = pd.DataFrame(y_pred , index = df_test.index , columns=['Target'])
result.to_csv("/kaggle/working/submission.csv")
```
