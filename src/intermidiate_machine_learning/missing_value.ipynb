```python
import pandas as pd 
from sklearn.model_selection import train_test_split
```

```python
data_set = pd.read_csv("../data/melb_data.csv" , index_col=0)
y = data_set.Price
X = data_set.drop("Price" , axis=1)
X = X.select_dtypes(exclude="object")
```

```python
X_train , X_valid , y_train , y_Valid = train_test_split(X , y , train_size=0.8, test_size=0.2,random_state=0)
```

```python
columns_with_missing = X_train.isnull().sum()
print(columns_with_missing[columns_with_missing > 0])
```

```python
from sklearn.impute import SimpleImputer

X_train_plus = X_train.copy()
X_valid_plus = X_valid.copy()


for col in columns_with_missing[columns_with_missing > 0].keys():
    X_train_plus[col + '_was_missing'] = X_train_plus[col].isnull()
    X_valid_plus[col + '_was_missing'] = X_valid_plus[col].isnull()

myImputer = SimpleImputer()
impute_X_train_plus = pd.DataFrame(myImputer.fit_transform(X_train_plus))
impute_x_valid_plus = pd.DataFrame(myImputer.transform(X_valid_plus))

impute_X_train_plus.columns = X_train_plus.columns
impute_x_valid_plus.columns = X_valid_plus.columns
```
